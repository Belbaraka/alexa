## Steered-Beamformer-Based Locators

The basic idea of this first approach is to sum up the contribution of each microphone afterappropriate filtering and look for a direction which maximize this sum.  This is referred asadelay-and-sum beamformer, it applies time shifts to the array signals to compensatefor the propagation delays.  Finding the direction that maximizes the aforementioned sumrequires solving anonlinear optimization problem.  Iterative searching methods such asstochastic gradient descent, orNewton-Raphsonare highly sensitive to initial position due toseveral local maximas

```python
import pyroomacoustics as pra
import numpy as np
import matplotlib.pyplot as plt

# Create a 8 by 5 metres room
room = pra.ShoeBox([8,5])

# Add microphones
R = pra.linear_2D_array([7.5, 1], 4, 0, 0.04)
room.add_microphone_array(pra.Beamformer(R, room.fs))

# Add a source 
room.add_source([2, 4])

# Compute the delay and sum weights for the beamformer
room.mic_array.rake_delay_and_sum_weights(room.sources[0][:1])

# plot the room and resulting beamformer
room.plot(freq=[1000, 2000, 4000, 8000], img_order=0)
plt.show()
```
