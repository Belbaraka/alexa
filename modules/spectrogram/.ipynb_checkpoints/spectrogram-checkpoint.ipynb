### ANALYSIS OF THE SPECTROGRAM FOR DIFFERENT AUDIO SIGNALS

```python
import numpy as np 
from scipy.io import wavfile
import IPython
import matplotlib.pyplot as plt
import pyroomacoustics as pra
```

##### Compute the spectrogram of an audio file 

```python
def compute_spectrogram(audio_file):
    fs, signal = wavfile.read(audio_file)
    
    signal_left = signal[:,0]
    signal_right = signal[:,1]
    
    fft_len = 512 
    block_size = fft_len - 50 + 1
    hop = block_size // 2
    
    l= signal.shape[0] // hop - 1
    
    
    fig = plt.figure()
    fig.set_size_inches(20,8)
    plt.subplot(2,1,1)
    plt.specgram(signal_left.astype(np.float32), NFFT=256, Fs=fs, vmin=-20, vmax=30) # dBFS scale
    plt.title('Stereo Left Signal', fontsize=22)
    plt.subplot(2,1,2)
    plt.specgram(signal_right.astype(np.float32), NFFT=256, Fs=fs, vmin=-20, vmax=30) # dBFS scale
    plt.title('Stereo Right Signal', fontsize=22)
```

```python
compute_spectrogram("male1.wav")
```
